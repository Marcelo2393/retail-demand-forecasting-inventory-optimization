# Se generaron variables temporales, lags y estadísticas móviles para capturar patrones de tendencia, estacionalidad y dependencia temporal en la demanda.

```python
import pandas as pd

df = pd.read_csv("/Users/marcelocallao/Desktop/Proyectos GitHub/Sistema de Predicción de Demanda y Optimización de Inventario en Retail/data/raw/retail_sales.csv", parse_dates=["date"])
df = df.sort_values(["store_id", "item_id", "date"])
```

```python
# Estacionalidad y comportamiento semanal
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day
df["dayofweek"] = df["date"].dt.dayofweek
df["weekofyear"] = df["date"].dt.isocalendar().week.astype(int)
```

```python
lags = [1, 7, 14, 28]

for lag in lags:
    df[f"lag_{lag}"] = df.groupby(["store_id", "item_id"])["sales"].shift(lag)
```

```python
# Rolling Feature
df["rolling_mean_7"] = df.groupby(["store_id", "item_id"])["sales"].shift(1).rolling(7).mean()
df["rolling_mean_14"] = df.groupby(["store_id", "item_id"])["sales"].shift(1).rolling(14).mean()
```

```python
# Rolling Std ( variablidad)
df["rolling_std_7"] = df.groupby(["store_id", "item_id"])["sales"].shift(1).rolling(7).std()
```

```python
# Eliminación de nulos
df = df.dropna()
```

```python
# Guardamos el dataset listo
df.to_csv("/Users/marcelocallao/Desktop/Proyectos GitHub/Sistema de Predicción de Demanda y Optimización de Inventario en Retail/data/processed/retail_sales_features.csv", index=False)
```
